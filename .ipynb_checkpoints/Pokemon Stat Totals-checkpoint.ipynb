# Pokemon Raw Stats Analysis
The objective of this project is to analyze the raw stats of pokemon from the popular Pokemon franchise. Our primary goal is to discover whether or not two separate factors can effect an overall pokemon specie's power.

## Dataset
The dataset we are using is a pre-formatted dataset of pokemon species and their raw stats. The dataset itself was found through Kaggle.com, which pulled its sources from multiple official pokemon sites for the raw statistics. The dataset includes species IDs, name, types, and all of the individual base stats.

```python
#importing pandas for using dataframes.
import pandas as pd

#importing in our locally available file.
data_path = "datasets/Pokemon.csv"
Pokemon = pd.read_csv(data_path)

#see the head of our dataframe so we know it is imported in properly
Pokemon.head()
```

## Pokemon Base Stats
All pokemon species have a set of base stats, with one corresponding to each stat used in the game. These stats are HP, Attack, Defense, Special Attack, Special Defense, and Speed. The base stats aree used as one of the primary numbers during a given pokemon's stat claculation. Though other factors can effect a given pokemon's statistics, they are consistent across all species.

# Cleaning the Dataset
The dataset we have is already quite clean, however we have two types of entries that we will be removing. The first is mega evolutions. These were all introduced as a part of generation 6 and are, by their nature, empowered versions of other species. This would skew the data immensely towards favoring generation 6. The other type of species we will be ignoring are legendary pokemon. These pokemon are also more powerful by their nature, as they are only attainable once per save file in the pokemon games. As certain generations have many more legendaries than others, this would also skew our data.

## Removing Mega evolutions
There is not a marker in the dataset for whether or not the species is its mega version. However, the mega version is always contains the substring 'Mega' within it. As such, we can create a mask for Mega evolutions and then slice the dataframe down.

```python
#We attempt to find the appropriate substring. If it is not found, it will return -1. Thus, a value of -1 means it is not a mega
mega_mask = Pokemon["Name"].str.find("Mega") == -1

#Apply the mask to a new variable to fully remove the mega evolutions
no_megas = Pokemon[mega_mask]

#displaying it to confirm the removal is complete
no_megas.head(10)
```

## Removing Legendaries
Removing legendaries is the simple task of creating a mask using the already boolean Legendary column

```python
#create the mask for being not legendary
leg_mask = no_megas["Legendary"] == False

#apply the mask
Pokemon_cleaned = no_megas[leg_mask]

#using tail to confirm since the last few in the original dataset are all legendary pokemon
Pokemon_cleaned.tail()
```

## Confirming Cleanliness
Here we will do a final few checks for null values to make sure that all of our data is in place.

```python
Pokemon_cleaned.isnull().sum()
```

The only null values we have is in the Type 2 column, which is understandable because that accounts for the mono-type pokemon. These rows will not be dropped.

# Hypotheses
This project aims to test two hypotheses, both of which are based upon the individual power of a pokemon species. All hypotheses are tested at a significance level of $p = 0.05$.

## Hypothesis 1
Is the generation a pokemon species from effect its base stat total?

$H_{0}:$ Species generation is not associated with base stat total

$H_{1}:$ Species generation is associated with base stat total

### Initial Graph Anlysis
We will be starting by graphing the generation of pokemon and the stat total as a boxplot to see if there is an obvious correlation.

```python
#We will be using the seaborn library for data visualisation
import seaborn as sns

sns.boxplot(x="Generation", y="Total", data=Pokemon_cleaned)
```

There seems to be variance, but no linear similarity between the two. As such, an analysis of variance, or f-test, will need to be run. This will check to see if there is a significant variance between the various generations of pokemon.

```python
#importing our statistics library of choice, scipy.stats
from scipy import stats

#set up a function for returning just the series for a given generation's total stats
def total_gen_stats(dataframe, generation):
    gen_mask = dataframe["Generation"] == generation #the mask for the gneration we are searching
    gen_slice = dataframe[gen_mask] #slices the dataframe to just the generation we are searching
    return gen_slice["Total"] #returns the total stats series for that slice

#initialize the variables for the six generations
gen1 = total_gen_stats(Pokemon_cleaned, 1)
gen2 = total_gen_stats(Pokemon_cleaned, 2)
gen3 = total_gen_stats(Pokemon_cleaned, 3)
gen4 = total_gen_stats(Pokemon_cleaned, 4)
gen5 = total_gen_stats(Pokemon_cleaned, 5)
gen6 = total_gen_stats(Pokemon_cleaned, 6)

#run our f-test, assigning the variables
f_stat, p_val = stats.f_oneway(gen1, gen2, gen3, gen4, gen5, gen6)

#displaying our p value
print("Our p value is: " + str(p_val))
```

### Result
Our p value is less than our significance level of 0.05, and thus we reject our null hypothesis to prove that there is an association between a pokemon species' generation and its base stat total.

## Hypothesis 2
Does the typing of a pokemon effect its base stat total?

$H_{0}:$ Pokemon typing is not associated with base stat total

$H_{0}:$ Pokemon typing is associated with base stat total

### Masking Our Data
Currently, the data says what type a given pokemon species is as one or two strings. As this is split into multiple columns and has several different values, we will be creating a function to facilitate the creation of our masks.

Additionally, the column being masked for is added as a column so we can graph our data more easily.

```python
#defining our masking function
def type_mask(typ):
    #create two separate masks for each of the primary and secondary type
    mask1 = Pokemon_cleaned["Type 1"] == typ
    mask2 = Pokemon_cleaned["Type 2"] == typ
    
    masked_df = Pokemon_cleaned[mask1|mask2] #logical operator to combine the masks
    
    masked_df_notated = masked_df.assign(Type = typ) #assign the extra column
    
    return masked_df_notated

#now to instantiate many different masks
grass_mask = type_mask("Grass")
fire_mask = type_mask("Fire")
water_mask = type_mask("Water")
bug_mask = type_mask("Bug")
normal_mask = type_mask("Normal")
poison_mask = type_mask("Poison")
electric_mask = type_mask("Electric")
ground_mask = type_mask("Ground")
fairy_mask = type_mask("Fairy")
fighting_mask = type_mask("Fighting")
psychic_mask = type_mask("Psychic")
rock_mask = type_mask("Rock")
ghost_mask = type_mask("Ghost")
ice_mask = type_mask("Ice")
dragon_mask = type_mask("Dragon")
dark_mask = type_mask("Dark")
steel_mask = type_mask("Steel")
flying_mask = type_mask("Flying")
```

### Graphing our data
Graphing our data is a little bit tricky, but doable. We added the type we masked for to its own column, which gives us the categorical value for us to graph by. We'll be concatenating all of these masked dataframes into a single daataframe that we can then graph a box plot by.

```python
#list of all of our masks
masks = [grass_mask, fire_mask, water_mask, bug_mask, normal_mask, poison_mask, electric_mask, ground_mask, fairy_mask, fighting_mask, psychic_mask, rock_mask, ghost_mask, ice_mask, dragon_mask, dark_mask, steel_mask, flying_mask]

#concatenate all of our masks together
concatenated = pd.concat(masks)

#import matplotlibs pyplot library for cleaning up our graph
import matplotlib.pyplot as plt

#sets the size for our figure to be read more easily
fig, ax = plt.subplots(figsize=(13, 3))

#sets the colors to each of the individual types for better visualization
type_colors = { "Normal": '#A8A77A',
                "Fire": '#EE8130',
                "Water": '#6390F0',
                "Electric": '#F7D02C',
                "Grass": '#7AC74C',
                "Ice": '#96D9D6',
                "Fighting": '#C22E28',
                "Poison": '#A33EA1',
                "Ground": '#E2BF65',
                "Flying": '#A98FF3',
                "Psychic": '#F95587',
                "Bug": '#A6B91A',
                "Rock": '#B6A136',
                "Ghost": '#735797',
                "Dragon": '#6F35FC',
                "Dark": '#705746',
                "Steel": '#B7B7CE',
                "Fairy": '#D685AD',}

#with the concatenated dataframe, we can now graph our boxplot
sns.boxplot(x="Type", y="Total", data=concatenated, ax=ax, palette=type_colors)
```

### Testing our Hypothesis
Similarly with our first hypothesis, we will be using an ANOVA f-test in order to test for significance.

```python
#running the f-test and assigning our variables 
f_stat, p_val = stats.f_oneway(grass_mask["Total"], fire_mask["Total"], water_mask["Total"], bug_mask["Total"], normal_mask["Total"], poison_mask["Total"], electric_mask["Total"], ground_mask["Total"], fairy_mask["Total"], fighting_mask["Total"], psychic_mask["Total"], rock_mask["Total"], ghost_mask["Total"], ice_mask["Total"], dragon_mask["Total"], dark_mask["Total"], steel_mask["Total"], flying_mask["Total"])

#displaying our p value
print("Our p value is: " + str(p_val))
```

### Result
Our p value is less than our significance level of 0.05, thus we reject our null hypothesis and show that there is an association between pokemon type and stat totals.

# Conclusions
We have rejected both of our null hypotheses, proving that both type and generation have an association with stat total indivdually. Though these can be seen subtly on the graphs plotted, it is proven by our statistical tests by their p values.
